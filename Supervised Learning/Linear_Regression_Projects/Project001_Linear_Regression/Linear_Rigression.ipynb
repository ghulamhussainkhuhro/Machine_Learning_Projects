# Predicting per capita income through linear Regression

### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

### Load the data from a CSV file

```python
df = pd.read_csv(r"C:\Users\Administrator\Machine_Learning_Projects\Supervised Learning\Linear_Regression_Projects\Project001_Linear_Regression\Data_Sets\canada_per_capita_income.csv")
```

```python
df.head()
```

### Method 1: Visualize the data using a scatter plot

```python
fig, ax = plt.subplots()

ax.scatter(df['year'], df['per capita income (US$)'])

plt.show()
```

### Method 1: Visualize the data using a scatter plot

```python
%matplotlib inline
plt.xlabel('Time in years')
plt.ylabel('Price in Dollars $')
plt.scatter(df['year'], df['per capita income (US$)'], color = 'r', marker = '+')
```

### Create a LinearRegression model

```python
reg =  linear_model.LinearRegression()
```

### Train the model
### Here, we are using 'year' as the input feature (X) and 'per capita income (US$)' as the target variable (y)

```python
reg.fit(df[['year']], df[['per capita income (US$)']])
```

### Get and print the coefficient and intercept of the trained model


```python
print(f"Coefficient (slope): {reg.coef_[0][0]}")
print(f"Intercept (y-intercept): {reg.intercept_[0]}")
```

### Predicting the per capita income for the year 2020
##### Make sure the input for predict is 2D by reshaping it

```python
predicted_income = reg.predict(pd.DataFrame([[2020]], columns=['year']))
```

### Print the predicted income for the year 2020

```python
print(f"Predicted per capita income for the year 2020 is: {predicted_income[0][0]} US$")
```

### Visualize the line of best fit

```python
plt.plot(df['year'], reg.predict(df[['year']]), color='blue')
plt.show()
```

### visualizing the actual value with the repdicted ones

```python
plt.plot(df['year'], reg.predict(df[['year']]), color='blue')
plt.scatter(df['year'], df['per capita income (US$)'], color='r', marker='+')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.show()
```
